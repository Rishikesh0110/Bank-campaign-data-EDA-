we have to numerical columns balance and age to handle outliers

**handling outlier for column age**

```python
#Ploting boxplot
plt.figure(figsize=(4,2))
sns.boxplot(bank_df["age"])
```

```python
#Finding inter quartile range
q1=bank_df["age"].quantile(0.25)
q3=bank_df["age"].quantile(0.50)
iqr=q3-q1
iqr
```

```python
#Caculating minimum whisker
minimum_whisker=q1-1.5*iqr
minimum_whisker
```

```python
#Caculating maximum whisker
maximum_whisker=q3+1.5*iqr
maximum_whisker
```

Observations
1. Value above 70 are considered as extreme value going to replece with the maximum whisker value

```python
#Subtituting extreme values with the median of column age
bank_df.loc[bank_df["age"] > 70,"age"] = bank_df["age"].median()
```

```python
#Again ploting box plot
plt.figure(figsize=(4,2))
sns.boxplot(bank_df["age"],orient="h",color="gold")
```

Observation
1. Now age column has no extreme values
